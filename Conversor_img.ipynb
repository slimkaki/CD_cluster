```python
# codigo que converte imagem em bytes
with open("img.jpg", "rb") as imageFile:
    f = imageFile.read()
    b = bytearray(f)

print(b)
```

```python
# codigo que converte imagem em base 64
import base64

with open("img.jpg", "rb") as imageFile:
    str = base64.b64encode(imageFile.read())
    print (str)
```
